```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from numpy.linalg import inv
```

```python
def N(mean, sigma):
    u = np.random.uniform(0, 1)
    v = np.random.uniform(0, 1)
    x = np.sqrt(-2*np.log(u)) * np.cos(2*np.pi*v)
    return mean + sigma * x
```

```python
u = np.random.uniform(0, 1, 10000)
v = np.random.uniform(0, 1, 10000)
x = np.sqrt(-2*np.log(u)) * np.cos(2*np.pi*v)

mean = 50
sigma = 3
num = x * sigma + mean
print('mean = %.2f'% num.mean())
print('sigma = %.2f'% num.std())
```

```python
plt.style.use('ggplot')
plt.hist(num, bins=100, align='mid')
plt.show()
```

```python
def seq_E(mean, sigma):
    print("Data point source function: N(%f, %f)\n" % (mean, sigma))
    seq = []
    seq.append(N(mean, sigma))
    tol = 0.1
    while tol > 0.01:
        point = N(mean, sigma)
        print("Add data point: %f" % point)
        seq.append(point)
        seq_arr = np.array(seq)
        print("Mean = %f  Variance = %f" % (seq_arr.mean(), seq_arr.std()))
        tol = max(abs(mean - seq_arr.mean()), abs(sigma - seq_arr.std()))
```

```python
seq_E(3.0, 5.0)
```

```python
def linear_data_generator(degree, W, sigma):
    rand_n = np.random.uniform(-1, 1)
    X = np.array([pow(rand_n, i) for i in range(degree)]).reshape(1, degree)
    e = N(0, sigma)
    return X, np.dot(X, W) + e
```

```python
a = 3
n = 3
b = 1
W = np.array([1, 2, 3]).reshape(n, 1)
S = b * np.identity(n)
m = np.zeros((n, 1))
tol = 999
while tol > 1e-6:
    X, y = linear_data_generator(n, W, a)
    print("\nAdd data point (%.5f, %.5f)" % (X[:, 1], y))
    v = a * np.dot(X.T, X) + S
    u = np.dot(inv(v), (a * np.dot(X.T, y) + np.dot(S, m)))
    print("\nPosterior mean:")
    print(u)
    print("\nPosterior variance:")
    print(inv(v))
    print("\nPredictive distribution ~ N(%.5f, %.5f)" % (np.dot(X, u), a + np.dot(np.dot(X, inv(v)), X.T)))
    print("-----------------------------------------")
    tol = abs((m-u).sum())
    S = v
    m = u
```

```python
a, n, b = 1, 4, 1
W = np.array([1, 2, 3, 4]).reshape(n, 1)

fig, ax = plt.subplots(2, 2, figsize=(8, 6))
result = np.array([[0, 500], [10, 50]])
title_names = [["Ground truth", "Predict result"], ["After 10 incomes", "After 50 incomes"]]

for row, vector in enumerate(result):
    for col, incomes in enumerate(vector):
#         print(row, col, incomes)
        px = []
        py = []
        S = b * np.identity(n)
        m = np.zeros((n, 1))
        for i in range(incomes):
            X, y = linear_data_generator(n, W, a)
            px.append(X[:, 1])
            py.append(y[:, 0])
            v = a * np.dot(X.T, X) + S
            u = np.dot(inv(v), (a * np.dot(X.T, y) + np.dot(S, m)))
            S = v
            m = u

        xx = np.linspace(-2, 2, 100)
        yy = []
        for value in xx:
            xx_vec = [pow(value, i) for i in range(n)]
            yy.append(xx_vec)
        yy = np.array(yy).reshape(-1, n)
        
        if incomes == 0:
            m = W
            S = 10000 * np.identity(n)
#             print(S)
        output = np.dot(yy, m)
        sigma = np.array([(a + np.dot(np.dot(yy, inv(S)), yy.T))[i, i] for i in range(100)]).reshape(-1, 1)
        
        ax[row, col].set_title(title_names[row][col])
        ax[row, col].scatter(px, py, s=5, c='b')
        ax[row, col].plot(xx, output, 'k-')
        ax[row, col].plot(xx, output+sigma, 'r-')
        ax[row, col].plot(xx, output-sigma, 'r-')
        ax[row, col].set_xticks(np.linspace(-1, 1, 3))
        ax[row, col].set_yticks(np.linspace(-10, 20, 4))
        ax[row, col].set_ylim(-15, 25)
        ax[row, col].set_xlim(-2, 2)
```
